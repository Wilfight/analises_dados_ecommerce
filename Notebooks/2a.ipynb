```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Extração

```python
df = pd.read_csv("bases/olist_orders_dataset.csv").drop_duplicates('order_id')
pedidos = df[df['order_status'] == 'delivered']
pedidos.head(10)
```

```python
pedidos_clone = pedidos[['order_id', 'order_approved_at', 'order_delivered_customer_date']]
pedidos_clone.head()
```

```python
pedidos_clone['order_approved_at'] = pd.to_datetime(pedidos_clone['order_approved_at'])
pedidos_clone['order_delivered_customer_date'] = pd.to_datetime(pedidos_clone['order_delivered_customer_date'])

diferencas_dias = pedidos_clone.groupby('order_id').apply(
    lambda d: (d['order_delivered_customer_date'] - d['order_approved_at']).dt.days)

diferencas_dias_df = diferencas_dias.reset_index(name='diferenca_dias')
diferencas_dias_df = diferencas_dias_df.drop(columns=['level_1'])
diferencas_dias_df.head()
```

```python
soma_dias = diferencas_dias_df['diferenca_dias'].sum()
print(soma_dias)
for i in diferencas_dias_df.columns[0:1].tolist():
    total_orders=len(diferencas_dias_df[i].astype(str).value_counts())
    print(total_orders)
    media= soma_dias/total_orders
    print("RESULTADO:")
    print("Media do tempo de entrega:",f'{media:.2f}',"dias")
```
